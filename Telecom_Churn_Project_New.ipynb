```python
columns_to_check = ['loc_og_t2o_mou', 'std_og_t2o_mou', 'loc_ic_t2o_mou']

threshold = 20

for column in columns_to_check:
    missing_percentage = df[column].isnull().mean() * 100
    print(f"Percentage of missing values in {column}: {missing_percentage:.2f}%")

    if missing_percentage <= threshold:
        mode_value = df[column].mode()[0]
        df[column].fillna(mode_value, inplace=True)
        print(f"Imputed missing values in {column} with mode: {mode_value}")
    else:
        df.drop(columns=[column], inplace=True)
        print(f"Deleted column {column} due to high percentage of missing values")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.impute import SimpleImputer


x = df_dummies.drop(columns=['churn'])
y = df_dummies['churn']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)


for df in [x_train, x_test]:
    for col in df.select_dtypes(include=['object', 'datetime']).columns:
        try:
            df[col] = pd.to_datetime(df[col])
            df[col + '_day'] = df[col].dt.day
            df[col + '_month'] = df[col].dt.month
            df[col + '_year'] = df[col].dt.year
            df.drop(columns=[col], inplace=True)
        except ValueError:
            df = pd.get_dummies(df, columns=[col], drop_first=True, dummy_na=False)


    imputer = SimpleImputer(strategy='median')
    numerical_cols = df.select_dtypes(include=np.number).columns
    df[numerical_cols] = imputer.fit_transform(df[numerical_cols])


logreg = LogisticRegression(class_weight='balanced')


param_grid = {
    'C': [0.1, 1, 10, 100],
    'solver': ['lbfgs', 'liblinear']
}


grid_search = GridSearchCV(logreg, param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)

best_model = grid_search.best_estimator_
y_pred_proba = best_model.predict_proba(x_test)[:, 1]
y_pred = best_model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)
```

```python
print(f"Best parameters: {grid_search.best_params_}")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
print(f"ROC AUC Score: {roc_auc}")
```
